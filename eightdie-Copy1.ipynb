```python
from qiskit import *
def random_eight_dice():
    #define quantum circuits
    qc = QuantumCircuit(15, 15)
    #spin those qbits
    qbit_measure_list = []
    
    for i in range(15):
        qc.h(i)
        qbit_measure_list.append(i)
        
    qc.measure(qbit_measure_list, qbit_measure_list) 
    
    print(qc.draw())
    def run_classical():
        %matplotlib inline
        simulator = Aer.get_backend('qasm_simulator')
        #122880 total qbits
        shots = 1024
        result = execute(qc, backend = simulator, shots=shots, memory=True).result()
        memory = result.get_memory()
        outputArray = []
        
        joined_qbits = "".join(memory)
        totalqbits = sum(len(i) for i in joined_qbits)
        print(totalqbits)
        return memory
    run_classical()

random_eight_dice()
```
